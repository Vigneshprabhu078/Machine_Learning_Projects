## **Sonar Rock vs Mine Predictions Project**

Our project focuses on developing a cutting-edge predictive model using sonar technology to differentiate between underwater ***rocks*** and ***mines***. By harnessing machine learning algorithms and extensive sonar signal data, we aim to enhance naval safety and operational efficiency. Stay tuned as we innovate in the realm of underwater security and defense.

**Importing the Dependencies**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

 **Data Collection and Data Preprocessing**

```python
#loading The Sonar data to pandas DataFrame
sonar_data=pd.read_csv('/content/sonar data.csv', header=None)
```

```python
#To Display First 5 dataset
sonar_data.head()
```

```python
# To check no of rows and columns
sonar_data.shape
```

```python
#statistical measure of data
sonar_data.describe()
```

```python
#To Check how many rock and mines in dataset
sonar_data[60].value_counts()
```

M --> Mine

R --> Rock

```python
sonar_data.groupby(60).mean()
```

```python
#Separating data and labels
X=sonar_data.drop(columns=60, axis=1)
Y=sonar_data[60]
```

```python
print(X) # x is data
print(Y) #y is label
```

**Training and Testing the data**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1) #test_size -0.1 -10 percentage of test data
                                                                              #stratify -- take equal number of Rock and Mine data's in Y
```

```python
#to check data's
print(X.shape,X_train.shape,X_test.shape)
```

```python
# training data and training label
print(X_train)
print(Y_train)
```

**Model Training**

**Logistic Regression**

```python
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

**Model Evaluation**

```python
# accuracy on  traing data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('Accuracy on training data : ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy on test data : ',test_data_accuracy)
```
