# **Predicting Calories Burned: A Machine Learning Approach using XGBoost Regressor**  -Vignesh Prabhu


In this project, we employed the XGBoost regressor to predict calories burned during physical activities. Using a dataset that included various activity metrics such as heart rate, duration, and intensity, we trained the model to accurately estimate calorie expenditure. By leveraging XGBoost's robustness in handling complex relationships and feature interactions, our goal was to provide a reliable tool for individuals and fitness enthusiasts to monitor and optimize their physical exertion levels.

**Import Dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

**Data Collection And Preprocessing**

```python
#Loading the dataset into DataFrame
calories= pd.read_csv('/content/calories.csv')
exercise= pd.read_csv('/content/Excercise.csv')
```

```python
# To check First 5 data's in dataset
calories.head()
```

```python
exercise.head()
```

**Combining the 2 DataFrames**

```python
calories_data= pd.concat([exercise,calories['Calories']],axis=1)
```

```python
calories_data.head()
```

```python
# Checking Number of Rows and Columns
calories_data.shape
```

```python
#Getting Complete Info About the data
calories_data.info()
```

```python
#Checking Null values
calories_data.isnull().sum()
```

**Data Analysis**

```python
#Statistical Measures
calories_data.describe()
```

**Data Visualization**

```python
sns.set()
```

**Correlation**

```python
# Convert 'Gender' column to numerical representation
calories_data['Gender'] = calories_data['Gender'].map({'male': 0, 'female': 1})

# Now calculate the correlation
Correlation = calories_data.corr()
print(Correlation)
```

```python
#Constructing HeatMap
plt.figure(figsize=(8,8))
sns.heatmap(Correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
#Converting Text data to Numerical Data
calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
```

```python
calories_data.head()
```

**Separating Feature and target**

```python
X= calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y= calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

**Spliting Data Into Training and testing Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

```python
model= XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

```python
#Prediction On Test Data
test_data_Prediction= model.predict(X_test)
```

```python
print(test_data_Prediction)
```

```python
#Mean Absolute Error
mae= metrics.mean_absolute_error(Y_test, test_data_Prediction)
print("Mean Absolute Error = ", mae)
```
